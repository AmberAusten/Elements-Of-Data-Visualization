```python
import altair as alt

from vega_datasets import data

iris = data.iris()
iris.head()
```

Using the Scatterplot Idiom we can show the relation between the Length and Width of pedals for each species.

```python
# Scatterplot 

alt.Chart(iris).mark_circle().encode(
    x='petalWidth:Q',
    y='petalLength:Q',
    color= alt.Color('species:N')
).interactive().properties(title="Petal Width vs Petal Length by Species")
```

From the bar chart we can compare the average petal length of each species.

```python
# Bar chart 

alt.Chart(iris).mark_bar().encode(
    x='mean(petalLength):Q',
    y='species:N',
    color = 'species',
).properties(title='Petal Length by Species')
```

```python
sp_means = iris.groupby('species').mean()
sp_means
```

This Radial visualization shows us the sepal length by species. This graphic is interesting because it shows variation within species, as well as how small the setosa species sepal length are compared to the other two




```python
#Radial 

base = alt.Chart(iris).encode(
    theta=alt.Theta("sepalLength:Q", stack = True),
    radius=alt.Radius("sepalLength", scale=alt.Scale(type="linear", zero=True, rangeMin=0)),
    color="species:N",
).properties(title="Sepal Length of Species")

c1 = base.mark_arc(innerRadius=20, stroke="#fff")
c1
```
